```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Cars.csv')
df
```

LINE PLOT 

```python
plt.plot(df['Selling_Price'],color='green',marker='o')
plt.title('SELLING PRICE')
plt.xlabel('Selling count')
plt.ylabel('Selling Prices')
plt.show()
```

SCATTERED PLOT

```python
plt.scatter(df['Kms_Driven'],df['Selling_Price'],color='red')
plt.title('Kms Driven vs Selling Price')
plt.xlabel('Selling Prices')
plt.ylabel('Distance in Kms')
plt.show()
```

BAR PLOT

```python
Value_Count = df['Seller_Type'].value_counts()

plt.bar('Dealer',Value_Count['Dealer'],color='red',label='DEALER')
plt.bar('Individual',Value_Count['Individual'],color='blue',label='INDIVIDUAL')
plt.title('Dealer vs Individual')
plt.xlabel('Seller Type')
plt.ylabel('Seller count')
plt.legend()
plt.show()
```

HISTOGRAM 

```python
plt.hist(df['Selling_Price'],color='red') # To see one value
plt.title('Histogram of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

POLAR PLOT

```python
plt.figure(figsize=(8,6))
ax = plt.subplot(111,projection='polar')
ax.scatter(df['Selling_Price'],df['Present_Price'])
```

BOX PLOT

```python
plt.boxplot(df['Selling_Price'])
plt.title('Histogram of Selling Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()# To see one value
```
