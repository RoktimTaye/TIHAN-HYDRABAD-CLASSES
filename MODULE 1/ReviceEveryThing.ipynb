```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Cars.csv')
df
```

```python
plt.plot(df['Selling_Price'],label = 'Selling Price')
plt.plot(df['Present_Price'],label = 'Present Price')
plt.legend()
plt.show()
```

```python
plt.scatter(df['Present_Price'],df['Selling_Price'])
plt.show()
```

```python
plt.bar(df['Fuel_Type'].value_counts().index,df['Fuel_Type'].value_counts())
plt.show()
```

```python
df['Present_Price'].value_counts().plot(kind='hist')
plt.legend()
plt.show()
```

```python
plt.hist(df['Selling_Price'],bins=10,edgecolor = 'black')
plt.show()
```

```python
df['Present_Price'].value_counts().plot(kind='box')
```

```python
fig,axs = plt.subplots(1, 3, sharex=True,figsize=(15, 5))
#1 row â†’ all plots will be placed horizontally.
#3 columns â†’ three separate plotting areas (called axes) are created side by side.
axs[0].boxplot(df['Selling_Price'])
axs[1].boxplot(df['Present_Price'])
axs[2].boxplot(df['Kms_Driven'])
plt.show()
```
