```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
MyDf = pd.read_csv('Cars.csv')
MyDf
```

```python
MyDf.describe()
```

```python
plt.Figure(figsize=(8,6))
plt.plot(MyDf['Present_Price'])
plt.plot(MyDf['Selling_Price'])
plt.xlabel('Price Category')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(MyDf['Selling_Price'], MyDf['Present_Price'])
plt.xlabel('Index')
plt.ylabel('Price')
plt.show()
```

```python
plt.bar(MyDf['Fuel_Type'].value_counts()
        .index,MyDf['Fuel_Type']
        .value_counts(),
        color=['green','blue','red'])
```

```python
plt.figure(figsize=(8,6))
ax = plt.subplot(111,projection='polar')
ax.scatter(MyDf['Present_Price'],MyDf['Kms_Driven'])
plt.show()
```

```python
fig,axs = plt.subplots(1, 3, sharex=True,figsize=(15, 5))
axs[0].boxplot(MyDf['Kms_Driven'])
axs[1].boxplot(MyDf['Present_Price'])
axs[2].boxplot(MyDf['Selling_Price'])
plt.show()
```

```python

plt.hist(MyDf['Selling_Price'],bins=5,edgecolor='black')
plt.show()
```

```python
MyDf2 = sns.load_dataset('tips')
CorrMatrix = MyDf2.corr(numeric_only=True)
sns.heatmap(CorrMatrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.show()
```

```python
sns.clustermap(CorrMatrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.show()
```

```python
numerical_cols = MyDf2.select_dtypes(include=['float64','int64']).columns
pairplot = sns.pairplot(MyDf2)
plt.show()
```

```python
numerical_cols = MyDf2.select_dtypes(include=['float64','int64']).columns
pairplot = sns.pairplot(MyDf2,
             hue='smoker',
             diag_kind='kde',
             kind='reg')
plt.show()
```

```python
sns.jointplot(data=MyDf2,
              x='total_bill',
              y='tip',
              kind='scatter',
              hue='sex')
plt.show()
```
