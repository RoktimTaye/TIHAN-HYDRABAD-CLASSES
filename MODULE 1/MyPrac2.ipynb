```python
df = sns.load_dataset('tips')
df
```

```python
#Normal Scattered plotting
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    size="size",
    hue="time",
    sizes=(20, 1000),
    alpha=0.6,
    edgecolor="w",
    linewidth=0.5
)
#Annotation for Highest point 
highest_bill_row = df.loc[df['total_bill'].idxmax()] #Storing max value in a variable for Annotating max point
plt.annotate(
    'Highest Bill',
    xy=(highest_bill_row['total_bill'],highest_bill_row['tip']), # **arrow annotating
    xytext=(highest_bill_row['total_bill']+5,highest_bill_row['tip']+1), # **Text annotating
    arrowprops=dict(facecolor='black',shrink=0.05),
    fontsize=10,
    color='red'
)
#Annotation for Lowest point
highest_bill_row = df.loc[df['total_bill'].idxmin()] #Storing min value in a variable for Annotating min point
plt.annotate(
    'Highest Bill',
    xy=(highest_bill_row['total_bill'],highest_bill_row['tip']), # **arrow annotating
    xytext=(highest_bill_row['total_bill']-5,highest_bill_row['tip']+0.5), # **Text annotating
    arrowprops=dict(facecolor='black',shrink=0.05),
    fontsize=10,
    color='green',
    horizontalalignment='right'
)
plt.show()
```
