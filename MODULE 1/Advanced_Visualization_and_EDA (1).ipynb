```python
#  Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

##  Seaborn Heatmap

```python
# Load Iris dataset
df = sns.load_dataset("iris")

# Correlation matrix
corr = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

##  Seaborn Pair-Plot

```python
# Pairplot
sns.pairplot(df, hue="species")
plt.suptitle("Pairplot of Iris Features by Species", y=1.02)
plt.show()
```

## Spotting Correlations & Outliers

```python
# Boxplot to detect outliers by category
sns.boxplot(data=df, x="species", y="sepal_length")
plt.title("Outlier Detection - Sepal Length by Species")
plt.show()
```

```python
# Boxplots for all numerical columns
df.drop(columns="species").plot(kind='box', subplots=True, layout=(2,2), figsize=(10,6), title="Boxplots for Outlier Detection")
plt.tight_layout()
plt.show()
```

##  Storytelling with Data

> "From the pairplot, we observe that *Setosa* species is well-separated from others, indicating good classification potential."

> "The heatmap shows a high positive correlation (0.87) between **petal length** and **petal width**, indicating potential multicollinearity."

##  Exercise: Penguins Dataset


```python
# Load penguins dataset
penguins = sns.load_dataset("penguins")
penguins.dropna(inplace=True)

# Pairplot
sns.pairplot(penguins, hue="species")
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(penguins.corr(numeric_only=True), annot=True, cmap="viridis")
plt.title("Penguins Dataset - Correlation Heatmap")
plt.show()
```
