```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df2 = sns.load_dataset('tips')
df2
```

```python
plt.Figure(figsize=(8,6))
Corr_Matrix = df2.corr(numeric_only=True)
sns.heatmap(Corr_Matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Corr of Numeric Features')
plt.show()
```

```python
Corr_Matrix = df2.corr(numeric_only=True)
sns.clustermap(Corr_Matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Corr of Most Related Features')
plt.show()
```

```python
numeric = df2.select_dtypes(include=['float64','Int64']).columns
pair_plot = sns.pairplot(df2,hue='smoker',
                         diag_kind='kde',
                         kind='reg')
pair_plot.figure.suptitle('Corr of numeric features',y=1.03)
plt.show()
```
