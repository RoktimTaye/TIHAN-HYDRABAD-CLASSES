```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Cars.csv')
df
```

LINE PLOT

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Selling_Price'], color='green',marker='o')
plt.xlabel('Index')
plt.ylabel('Selling Price')
plt.title('Selling Price line graph')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Selling_Price'], color='green',marker='o')
plt.plot(df['Present_Price'], color='blue', marker='x')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend(['Selling Price', 'Present Price'])
plt.title('Selling Price vs Present Price')
plt.show()
```

```python
plt.bar(df['Fuel_Type'].value_counts().index,df['Fuel_Type'].value_counts(),color=['red', 'blue', 'green'])
plt.title('Fuel Type Bar Graph')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
legends = ['Petrol', 'Diesel', 'CNG']
plt.legend(legends)
plt.show()
```

```python
plt.scatter(df['Selling_Price'], df['Present_Price'], color='green', marker='o')
plt.title('Selling Price vs Present Price Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend(['Selling Price', 'Present Price'])
plt.show()
```

```python
df.select_dtypes(include='number').hist(bins=5, edgecolor='black')
plt.show()
```

```python
plt.hist(df['Selling_Price'])
plt.title('Histogram of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
ax = plt.subplot(111, projection='polar')
ax.scatter(df['Selling_Price'], df['Kms_Driven'])  # Selling price as angle, km driven as radius
plt.title('Polar Plot: Selling Price vs Kilometers Driven')
plt.show()
```

```python
plt.boxplot(df['Selling_Price'])
plt.title('Box Plot of Selling Price')
plt.ylabel('Price')
plt.show()
```

```python
plt.boxplot([df['Selling_Price']], labels=['Selling Price'])
plt.boxplot([df['Present_Price']], labels=['Present Price'])
plt.title('Box Plots Comparison')
plt.ylabel('Price')
plt.xlabel('Both Selling Category')

```

```python
plt.scatter(df['Year'], df['Selling_Price'])
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.title('Dot Plot of Selling Price vs Year')
plt.show()
```
