```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

LINE PLOT

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Selling_Price'], color='green',marker='o')
plt.xlabel('Index')
plt.ylabel('Selling Price')
plt.title('Selling Price line graph')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Selling_Price'], color='green',marker='o')
plt.plot(df['Present_Price'], color='blue', marker='x')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend(['Selling Price', 'Present Price'])
plt.title('Selling Price vs Present Price')
plt.show()
```

```python
plt.bar(df['Fuel_Type'].value_counts().index,df['Fuel_Type'].value_counts(),color=['red', 'blue', 'green'])
plt.title('Fuel Type Bar Graph')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
legends = ['Petrol', 'Diesel', 'CNG']
plt.legend(legends)
plt.show()
```

```python
plt.scatter(df['Selling_Price'], df['Present_Price'], color='green', marker='o')
plt.title('Selling Price vs Present Price Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend(['Selling Price', 'Present Price'])
plt.show()
```

```python
df.select_dtypes(include='number').hist(bins=5, edgecolor='black')
plt.show()
```

```python
plt.hist(df['Selling_Price'])
plt.title('Histogram of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```
