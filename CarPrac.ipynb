```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df2 = pd.read_csv('MODULE 1\Cars.csv')
df2
```

```python
df2.describe()
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
plt.Figure(figsize=(8, 6))
plt.plot(df2['Selling_Price'], color='green', marker='o')
plt.plot(df2['Present_Price'], color='blue', marker='x')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend(['Selling Price', 'Present Price'])
plt.title('Selling Price vs Present Price')
plt.show()
```

```python
plt.scatter(df2['Selling_Price'], df2['Present_Price'], color='green', marker='o')
plt.title('Selling Price vs Present Price Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend(['Selling Price', 'Present Price'])
plt.show()
```
