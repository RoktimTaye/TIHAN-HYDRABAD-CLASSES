```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import pairplot

df = sns.load_dataset("tips")
display(df.head())
display(df.info())
```

```python
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm",fmt='g')
plt.show()
```

```python
numerical_cols = df.select_dtypes(include=['float64','int64']).columns
pair_plot = sns.pairplot(df[numerical_cols])
pair_plot.fig.suptitle('pair plot of numerical columns',y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex')
plt.title('total bill vs tip ny sex')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df,x='total_bill',y='tip')
plt.title('tip ampunt by sex')
plt.show()
```

```python
df = sns.load_dataset('tips')

# Create the bubble plot
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    size="size",
    hue="time",
    sizes=(20, 1000),
    alpha=0.6,
    edgecolor="w",
    linewidth=0.5
)
plt.title('Bubble Plot of Tips Data with Annotations')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend(title='Party Size / Time of Day', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

heighest_bill_row = df.loc[df['total_bill'].idxmax()]
plt.annotate('Heighest bill',
             xy=(heighest_bill_row['total_bill'], heighest_bill_row['tip']),
             xytext=(heighest_bill_row['total_bill'] + 5, heighest_bill_row['tip'] + 5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=11,
             color='red'
             )
lowest_bill_row = df.loc[df['total_bill'].idxmin()]
plt.annotate(
    'Lowest Bill!',
    xy=(lowest_bill_row['total_bill'], lowest_bill_row['tip']),
    xytext=(lowest_bill_row['total_bill'] - 7, lowest_bill_row['tip'] + 0.5), # Adjust text position
    arrowprops=dict(facecolor='green', shrink=0.05),
    fontsize=10,
    color='green',
    horizontalalignment='right' # Align text to the right of the arrow
)

plt.show()
```
