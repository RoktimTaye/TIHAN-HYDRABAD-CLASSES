```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df2 = pd.read_csv('MODULE 1/Cars.csv')
df2
```

```python
df2.describe()
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
plt.Figure(figsize=(10, 5))
plt.plot(df2['Selling_Price'],color='blue', label='Selling Price',marker='o')
plt.plot(df2['Present_Price'], color='red', label='Present Price',marker='x')
plt.legend()
plt.title('Selling Price vs Present Price')
plt.xlabel('Selling Price')
plt.ylabel('Present Price')
plt.show()
```

```python
plt.bar(df2['Fuel_Type'].value_counts().index,df2['Fuel_Type'].value_counts(),color=['red', 'blue', 'green'])
plt.title('Fuel Type Bar Graph')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
legends = ['Petrol', 'Diesel', 'CNG']
plt.legend(legends)
plt.show()
```

```python
plt.bar(df2['Seller_Type'].value_counts().index,df2['Seller_Type'].value_counts(),color=['orange', 'purple'])
plt.title('Seller Type Bar Graph')
plt.xlabel('Seller Type')
plt.ylabel('Count')
legends = ['Individual', 'Dealer']
plt.legend(legends)
plt.show()
```

```python
plt.scatter(df2['Selling_Price'], df2['Present_Price'], color='red', marker='o')
plt.title('Selling Price vs Present Price Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend(['Selling Price', 'Present Price'])
plt.show()
```
